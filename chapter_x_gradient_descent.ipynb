```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
param = np.linspace(-10, 10, 400)
loss = param**2  # Quadratic function as a simple loss function

# Calculate the gradient at two points
param1 = -5
param2 = 5
loss1 = param1**2
loss2 = param2**2
gradient1 = 2*param1
gradient2 = 2*param2

# Create a new plot with a title and axis labels
plt.figure(figsize=(10, 6))
plt.plot(param, loss, color = "navy", alpha=0.5, linewidth=3)

plt.xlabel(r'$\theta$')
plt.ylabel(r'$L_{M \vec{x} \vec{y}}(\theta)$')

# Remove top and right border
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


newline = '\n'
loss_string = r'$L_{M \vec{x} \vec{y}}(\theta)$'
derivitive_string = r'$\frac{\partial L_{M \vec{x} \vec{y}}(\theta)} {\partial \theta}$'
theta_string = r'$\theta$'
# Add points
plt.scatter([param1, param2], [loss1, loss2], color='firebrick')
plt.scatter([0, 0], [0, 0], color='firebrick')



plt.text(param1 - 2.0, loss1, f'{theta_string} =   {param1} {newline} {loss_string} =  {loss1} {newline} {derivitive_string} = {gradient1}', fontsize=15, ha='right')
plt.text(param2 + 5.5, loss2, f'{theta_string} =   {param2} {newline} {loss_string} = {loss2} {newline} {derivitive_string} = {gradient2}', fontsize=15, ha='right')

# Add arrows pointing towards 0 on the x-axis
plt.annotate('', xy=(param1/2, loss1), xytext=(param1 + 0.3, loss1 - 0.2), arrowprops=dict(facecolor='firebrick', edgecolor='none', shrink=0.05, width=5, headwidth=8, headlength = 8))
plt.annotate('', xy=(param2/2, loss2), xytext=(param2 - 0.3, loss2 - 0.2), arrowprops=dict(facecolor='firebrick', edgecolor='none', shrink=0.05, width=5, headwidth=8, headlength = 8))


# Show the plot
plt.show()
```
